### Desafio: lendo dados com vários parâmetros
1.   Verifique se o arquivo CSV está separado por vírgula ou ponto e vírgula.
2.   A codificação do arquivo é ISO-8859-1.
3. As três primeiras linhas linhas do arquivo podem ser desconsideradas, pois o cabeçalho só começa na quarta linha.
4. As 9 últimas linhas também podem ser desconsideradas, pois são apenas informações sobre onde os dados foram obtidos.
5. Para deletar as últimas linhas é necessário adicionar o parâmetro engine='python'.

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv'
dados_sus = pd.read_csv(url, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter = 9, engine='python')
dados_sus.head()
```

## Manipulando arquivos JSON
JavaScript Object Notation

```python
dados_pacientes = pd.read_json('https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes.json')
dados_pacientes
```

```python
dados_pacientes_2 = pd.read_json('https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes_2.json')
dados_pacientes_2
```

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
df_normalizado
```

```python
df_normalizado.to_json('historico_pacientes_normalizado.json')
```

### Desafio: normalizando o JSON obtido na API
O desafio agora é normalizar esse DataFrame.

```python
import requests
import json

dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
resultado = json.loads(dados_frutas.text)
```

```python
dados_frutas_normalizado = pd.json_normalize(resultado)
dados_frutas_normalizado
```
